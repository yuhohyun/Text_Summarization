```python
from bs4 import BeautifulSoup

# HTML 태그 제거 함수
def remove_html_tags(text):
    soup = BeautifulSoup(text, 'html.parser')
    return soup.get_text()

# 모든 셀에 대해 HTML 태그 제거 적용
#df_train = df_train.applymap(lambda x: remove_html_tags(str(x)))
df_val = df_val.applymap(lambda x: remove_html_tags(str(x)))
```
